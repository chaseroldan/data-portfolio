```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('completeanalysis.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
coke = df.loc[(df['Company Name'] == 'Coca-Cola')].sort_values(by='Company Name', ascending=True)
coke
```

```python
pepsi = df.loc[(df['Company Name'] == 'PepsiCo')].sort_values(by='Company Name', ascending=True)
pepsi
```

```python
combined = coke.append(pepsi, ignore_index = True)
combined
```

```python
totalrevenue = combined.plot.bar(x='Company Name', y='Total Revenue 2018 - 2020', rot=0)
plt.savefig('totalrevenue.png')
```

```python
totalnetincome = combined.plot.bar(x='Company Name', y='Total Net Income 2018 - 2020', rot=0)
plt.savefig('totalnetincome.png')
```

```python
totalprofitmagin = combined.plot.bar(x='Company Name', y='Total Profit Margin 2018 - 2020', rot=0)
plt.savefig('totalprofitmargin.png')
```

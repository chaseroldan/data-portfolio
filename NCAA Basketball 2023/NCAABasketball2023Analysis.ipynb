```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ncaabasketball2023.csv')
```

```python
df["PTS-RATIO"] = df["PTS"] / df["PTS-ALLOWED"]
```

```python
pctpts = df.plot.scatter(x='PTS', y='PCT')
plt.savefig('pctpts.png')
```

```python
pctreb = df.plot.scatter(x='REB', y='PCT')
```

```python
pctallowed = df.plot.scatter(x='PTS-ALLOWED', y='PCT')
plt.savefig('pctptsallowed.png')
```

```python
pctptsratio = df.plot.scatter(x='PTS-RATIO', y='PCT')
plt.savefig('pctptsratio.png')
```
